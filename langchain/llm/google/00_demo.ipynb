```python
base_url = 'https://generativelanguage.googleapis.com/v1'
proxy_url = 'https://gemini-api.funning.top/v1'
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
import os

load_dotenv()

google_api_base = os.getenv("GOOGLE_API_BASE")
google_api_key = os.getenv("GOOGLE_API_KEY")

print(google_api_key, google_api_base)

# https://gemini-api.funning.top/
llm = ChatGoogleGenerativeAI(
    model="gemini-1.5-pro",
    google_api_key=google_api_key,
    # client_options={'api_endpoint': proxy_url},
    temperature=0,
    max_tokens=None,
    timeout=None,
    max_retries=2,
)
```

```python
llm.invoke("你好，今天天气怎么样")
```

```python
from langchain_openai import ChatOpenAI

# qwen 兼容 openai的接口
gemini_llm_openai = ChatOpenAI(
    openai_api_key=os.getenv("GOOGLE_API_KEY"),
    openai_api_base=base_url,
    # model_name="gemini-1.5-flash",
    model_name="gemini-1.5-flash-latest",
    temperature=0,
    streaming=True,
)
```

```python
gemini_llm_openai.invoke("你好，今天天气怎么样")
```
