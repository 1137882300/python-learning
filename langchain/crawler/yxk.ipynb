```python
from langchain.prompts import PromptTemplate

template = """在 >>> 和 <<< 之间是网页的返回的HTML内容。

网页是游侠客旅游网站的一个页面，包含了一些旅游产品的信息。请从中抽取出：产品的名称、价格、描述、产品特色、出发地、行程的名称、批次的名称、以及每个批次日期对应的价格（需要关注是儿童还是成人的价格）、活动亮点、每日的具体行程、费用说明、适合年龄、活动装备、目的地介绍、游玩的天数。

请抽取参数请求的信息。 并用markdown 语法 作为输出，总之你需要做好格式化数据。

>>> {requests_result} <<<
请使用如下的markdown格式返回数据
| *   [武功山](http://www.youxiake.com/search/results/0-1696-0-0-0-0.html)<br>    <br>*   [花鸟岛](https://www.youxiake.com/search/results/0-24-0-0-0-0.html)<br>    <br>*   [国内海岛](http://www.youxiake.com/dahai/)<br>    <br>*   [徒步恩施](http://www.youxiake.com/search/results/0-1979-0-0-0-0.html)<br>    <br>*   [朝鲜](http://www.youxiake.com/search/results/0-5413-0-0-0-0.html)<br>    <br>*   [吴哥迷城](https://bbs.youxiake.com/mdd/8098.html)<br>    <br>*   [东欧](http://www.youxiake.com/search/results/0-5370-0-0-0-0.html)<br>    <br>*   [徒步雨崩](http://www.youxiake.com/lines.html?id=19756)<br>    <br>*   [越南全景](https://bbs.youxiake.com/mdd/1708.html)<br>    <br>*   [田园靖西](http://www.youxiake.com/lines.html?id=22510)<br>    <br>*   [厦门](http://www.youxiake.com/search/results/0-100-0-0-0-0.html)<br>    <br>*   [太行山](http://www.youxiake.com/search/results/0-2197-0-0-0-0.html)<br>    <br>*   [小众印尼](http://www.youxiake.com/search/results/0-0-0-0-0-0.html?keyword=小众印尼)<br>    <br>*   [摩洛哥](http://www.youxiake.com/search/results/0-681-0-0-0-0.html)<br>    <br>*   [澳大利亚](http://www.youxiake.com/search/results/0-2677-0-0-0-0.html)<br>    <br>*   [万塔之城-蒲甘](https://bbs.youxiake.com/mdd/850.html) |
| 国内  |
| *   [四川](javascript:;)<br>    <br>*   [云南](javascript:;)<br>    <br>*   [北京](javascript:;)<br>    <br>*   [东北](javascript:;)<br>    <br>*   [贵州](javascript:;)<br>    <br>*   [西藏](javascript:;)<br>    <br>*   [湖南](javascript:;)<br>    <br>*   [湖北](javascript:;)<br>    <br>*   [广西](javascript:;)<br>    <br>*   [西北](javascript:;)<br>    <br>*   [华南](javascript:;)<br>    <br>*   [福建](javascript:;)<br>    <br>*   [华东](javascript:;)<br>    <br>*   [秦晋](javascript:;)<br>    <br>*   [华北](javascript:;)<br>    <br>*   [华中](javascript:;)<br>    <br>*   [重庆](javascript:;)<br>    <br>*   [内蒙古](javascript:;)<br>    <br>*   [新疆](javascript:;) |
| 出境  |
| *   [东南亚](javascript:;)<br>    <br>*   [欧洲](javascript:;)<br>    <br>*   [南亚](javascript:;)<br>    <br>*   [日韩朝蒙](javascript:;)<br>    <br>*   [尼泊尔](javascript:;)<br>    <br>*   [越南](javascript:;)<br>    <br>*   [东欧](javascript:;)<br>    <br>*   [大洋洲](javascript:;)<br>    <br>*   [俄罗斯](javascript:;)<br>    <br>*   [泰国](javascript:;)<br>    <br>*   [美洲](javascript:;)<br>    <br>*   [斯里兰卡](javascript:;)<br>    <br>*   [东非](javascript:;)<br>    <br>*   [缅甸](javascript:;)<br>    <br>*   [印度尼西亚](javascript:;)<br>    <br>*   [摩洛哥](javascript:;) |

服务热线 9:00~21:00
 
站点提示

按照工信部社交网站实名制规定，需绑定手机号实名认证

[确定](https://www.youxiake.com/account/setting?check=2)

取消

确定

请先登录×

登录成功！

账号/密码错误！

登录中

 

[忘记密码？](https://www.youxiake.com/forget/passwrod)
 | [注册新账号](https://www.youxiake.com/reg?refer=https%3A%2F%2Fwww.youxiake.com%2Flines.html%3Fid%3D22510)
 | [帮助中心](https://www.youxiake.com/help)

###### 游侠客账号注册/登录

获取验证码

正在加载验证码...

一键注册/登录

账号密码登录>

未注册游侠客账号的手机，登录时将同时注册，登录代表您已同意 [《注册服务协议》](http://www.youxiake.com/corp/agreement)
和 [《隐私协议》](http://www.youxiake.com/corp/privacy)

提示内容

跟团游 · 生态乡野 · 小众风光 · 游山玩水

*   ![](https://gallery.youxiake.com/product/line/202404/11/1712822550916.jpg?imageView2/1/w/502/h/280/q/85)
*   ![](https://gallery.youxiake.com/product/line/202405/06/1714985050091.jpg?imageView2/1/w/502/h/280/q/85)
*   ![](https://gallery.youxiake.com/product/line/202212/06/1670314030239.jpg?imageView2/1/w/502/h/280/q/85)
*   ![](https://gallery.youxiake.com/product/line/202302/28/1677574150197.jpg?imageView2/1/w/502/h/280/q/85)
*   ![](https://gallery.youxiake.com/product/line/202405/06/1714984919668.jpg?imageView2/1/w/502/h/280/q/85)

*   日历查看
*   列表查看

*   日
*   一
*   二
*   三
*   四
*   五
*   六

田园靖西 | 广西崇左田园风光，德天跨国瀑布4日
------------------------

南宁集散 | 探索中国侏罗纪公园-通灵大峡谷 | 深入边陲西南小城-靖西 | 确保入住明仕景区内的明仕山庄

编号：GZ-22510

**8150** _人已报名_ \[复制\] 收藏 手机购买  ![田园靖西 | 广西崇左田园风光，德天跨国瀑布4日](https://www.youxiake.com/qrcode/22510.png) _手机扫一扫_ ![游侠客旅行](https://www.youxiake.com/qrcode/miniapp/22510.png) _微信扫码_

4天3晚

产品价格： **￥** **1780** 元起（成人） **￥** **950** 元起（儿童）

起价说明 **本起价是可选活动日期中，按单人出行核算的最低单人价格。产品价格会根据您所选活动日期、出行人数、游客类型以及所选附加项目的不同而有所差别。** 满减优惠***   活动时间：2024-01-10 18:00:00 - 2025-05-31 00:00:00
_满6人, 优惠50元/人_  
_满10人, 优惠100元/人_  
**

产品特色：

小众目的地 西南秘境 明仕篝火晚会 峡谷避暑

|     |     |
| --- | --- |
| 供 应 商： | [游侠客国际旅游有限公司广州分公司](https://www.youxiake.com/corp/zizhi/img?company_id=7) |
| 出 发 地： | 南宁  |
| 退订政策： | 支持退款 特殊原因退订保障 查看政策 > |
| 出团通知： | 周边游提前1天，国内游提前3天，出境游提前3-5天，APP和短信群发出团通知 |

游侠客 · 定制旅行 不跟团不自助，让旅行更省心 [开启定制之旅>>](https://www.youxiake.com/dingzhi)

*   [报名须知](https://www.youxiake.com/help/cate/35)
    
*   [优惠政策](https://www.youxiake.com/help/cate/91)
    

行程套餐：

*   入住明仕山庄酒店

_\*请选择行程套餐_

批次套餐：

*   南宁集散

_\*请选择批次套餐_

*   套餐：
*   
*   行程：
*   
*   批次：
*   
*   成人：
*   \-  +
    
    123456789101112131415161718192021222324252627282930313233343536373839404142434445464748495051
    
*   儿童：
*   \-  +
    
    123456789101112131415161718192021222324252627282930313233343536373839404142434445464748495051
    

总额： ￥ **0**

活动费用

立刻报名

可选资源

其他费用 费用说明 单价 数量
 
[立刻报名](#newDetail)

*   入住明仕山庄酒店

【入住明仕山庄酒店】 适用于以下批次套餐：

南宁集散

展开∨

产品经理说
-----

广西西南沿着中越边境，竟坐落着如此多遗世独立的秀美小城，这里媲美九寨！叫板恩施！还藏着惊艳世界的奇观！

一路，森林、峡谷、溶洞、瀑布、湖泊…震撼大气的美景不缺，又仙又野，4天3晚一起感受这座“又仙又野”的隐世小城。

活动亮点
----

**花样好评：**年收客2000+人，每团均花式好评，多年独立操作，放心报名

**小众目的地：**广西西南边境，游览西南边境风景极秀丽的景观

**通灵大峡谷：**走进清凉峡谷，落差188.6米的西南第一高瀑布，广西版“侏罗纪世界”的植被世界

**德天跨国瀑布：**亚洲第一、世界第四大跨国瀑布，独特的山水景色，成为多部仙侠剧的拍摄点

**渠洋湖游船：**被称为山中的大海，如喀斯特峰林中的一颗明珠，泛舟在“广西版下龙湾”景色中赏日落时刻

**西南名泉鹅泉：**西南地区三大名泉美称的鹅泉，五色泉水与别样的田园风光

**入住明仕山庄：**特别安排入住世外桃源明仕山庄，明仕田园景区核心位置，一步一景

**深度纯玩：**全程纯玩无购物，游侠客领队带队朋友式旅行

![田园靖西 | 广西崇左田园风光，德天跨国瀑布4日 - 南宁集散 | 探索中国侏罗纪公园-通灵大峡谷 | 深入边陲西南小城-靖西 | 确保入住明仕景区内的明仕山庄](http://gallery.youxiake.com/Public/Data/upload/productimg/202003/30/5e81c085bf6ce.jpg)

![田园靖西 | 广西崇左田园风光，德天跨国瀑布4日 - 南宁集散 | 探索中国侏罗纪公园-通灵大峡谷 | 深入边陲西南小城-靖西 | 确保入住明仕景区内的明仕山庄](https://gallery.youxiake.com/product/line/202404/26/1714103691155.jpg)

![田园靖西 | 广西崇左田园风光，德天跨国瀑布4日 - 南宁集散 | 探索中国侏罗纪公园-通灵大峡谷 | 深入边陲西南小城-靖西 | 确保入住明仕景区内的明仕山庄](https://gallery.youxiake.com/product/line/202405/07/1715046936401.jpg)

![田园靖西 | 广西崇左田园风光，德天跨国瀑布4日 - 南宁集散 | 探索中国侏罗纪公园-通灵大峡谷 | 深入边陲西南小城-靖西 | 确保入住明仕景区内的明仕山庄](https://gallery.youxiake.com/product/line/202406/14/1718344878983.jpg)

![田园靖西 | 广西崇左田园风光，德天跨国瀑布4日 - 南宁集散 | 探索中国侏罗纪公园-通灵大峡谷 | 深入边陲西南小城-靖西 | 确保入住明仕景区内的明仕山庄](https://gallery.youxiake.com/product/line/202404/26/1714102359041.jpg)

![田园靖西 | 广西崇左田园风光，德天跨国瀑布4日 - 南宁集散 | 探索中国侏罗纪公园-通灵大峡谷 | 深入边陲西南小城-靖西 | 确保入住明仕景区内的明仕山庄](https://gallery.youxiake.com/product/line/202405/06/1714981135633.jpg)

具体行程 重要提示：以下行程仅为参考，具体以实际签订合同为准
------------------------------

行程概要
====

天数：4天3 晚 出发：全国出发 到达：南宁 解散：南宁

D1

第1 天 全国

南宁★集合日

·全天

**全天南宁集合日。自行前往入住游侠客指定酒店。**

集合酒店位于市区，交通便利，早到的游侠可去周边游玩和寻觅美食，自由活动期间请妥善保管随身携带物品，注意安全！

  

自由活动推荐

南湖、青秀山、民族博物馆、三街两巷、南宁国际会展中心、航洋国际、南湖公园、青秀山景区等。

前往美食街区中山路，建政路夜市，当然少不了品尝当地的美食，如老友粉、手榨米粉、糖水等等。

  

集合酒店交通指引

南宁吴圩机场 ： 网约车约90元~120元， 或者乘机场大巴1号线至民航酒店下车，再打车至酒店（约10元）

南宁东站： 打车约50元，也可以坐地铁1号线到南宁火车站步行到酒店

南宁火车站：出站后步行1公里或者打车（约10元）

用餐：全天自理

住宿：南宁酒店标准双人间（参考南宁宜尚Plus酒店或雅斯特国际酒店或网评同级酒店）

备注：一般提前2~3天会建群联系

集结日：全天集合：南宁市区酒店标间

D2

第2 天 南宁

通灵大峡谷

渠洋湖

靖西

亮点： 打卡西南瀑布通灵大峡谷 渠洋湖游船赏日落

·全天

早餐后乘坐大巴前往通灵大峡谷**。**

车上相互自我介绍，领队介绍景区，并做一些有趣的互动小游戏，在欢声笑语中迎接接下来的旅途。

到达景区门口后我们先AA腐败解决午餐，随后跟领队一起进入景区。沿着景区的楼梯向下穿越溶洞，来到峡谷底，峡谷内水流潺潺，拥有着举世罕见的西南第一高瀑布—**通灵大瀑布**，同时峡谷内还拥有洞中瀑布、地下暗河、洞穴奇观、古石垒、古悬葬、原始植被等自然景观与人文景观。

适时集合，我们驱车前往渠洋湖，这里是一个喀斯特高原湖泊水库，峰林小岛星星点点散落在湖中，配合上清澈透蓝的湖水，拍个照片发到朋友圈，别人还以为你在下龙湾呢。将近日落时分，湖水倒影着夕阳，我们乘坐游船慢慢欣赏，惬意至极。

游览完毕后乘车入住靖西市区酒店。

车程：南宁-通灵大峡谷约3小时；通灵大峡谷-渠羊湖1.5小时；渠羊湖-靖西酒店约50分钟 用餐：含早，正餐自理，可AA腐败

住宿：靖西酒店标准双人间（参考：靖西雅斯特酒店或同级酒店）

备注：假期日期间，通灵大峡谷可能会有堵车的情况，如影响渠羊湖看日落的时间，领队有权调整行程为鹅泉景区，渠羊湖改为第二天的上午行程。  
游玩通灵大峡谷景区需要上下景区楼梯，下楼梯路段大概818级楼梯，上楼梯的路段相对平缓点，没有严重的膝盖问题均可参加。

D3

第3 天 靖西

鹅泉

德天大瀑布

明仕山庄

亮点： 名泉靖西鹅泉 德天跨国大瀑布

·全天

早餐后集合后驱车前往鹅泉

这里是一处面积达到50平方公里的巨泉，有西南三大名泉的美称，泉水由透明到绿色到蓝色，泉眼位置的水深达到了30米，德天瀑布的源头就是此处，我们可以在泉水边静静的发呆，也可以攀登到半山亭俯瞰鹅泉全景。

午餐后接着乘坐大巴前往德天跨国大瀑布

德天大瀑布，取得天独厚之意，一半属中国、一半属越南，一不小心就出国了。激流从70米高山崖飞泻而下，三层跌落如万马奔腾，轰鸣如水龙吟，《花千骨》长流山仙界的取景地。（节假日期间德天景区乘坐景区接驳车的人比较多，遇到排队的情况，请跟随团队耐心等候）

之后前往**明仕田园**

到达明仕田园后，我们将入住位于景区内的明仕山庄酒店，酒店被山水包围、诗情画意、颇有隐居仙境之意。游侠们可在此赏景、喝茶、漫步、发呆、闲聊。

晚上我们围着篝火晚会一起跳舞，和来自各地的游侠们一起感受热闹的夜晚。

车程：靖西-鹅泉 30分钟；鹅泉-德天瀑布约2小时；德天瀑布-明仕山庄约45分钟 用餐：含早，正餐自理

住宿：大新明仕山庄民族特色双床房（只住明仕山庄酒店，订不到房不发团！）

备注：骑行明仕田园，凭房卡可租自行车（会有比较老旧的情况），或可自费租用山地单车等。因有一段公路骑行，老人家与小朋友不建议参与到骑行中。  
明仕山庄酒店泳池政策：需购买168元/张的游泳券方可进入泳池（具体以酒店前台为准）  

D4

第4 天 明仕山庄

明仕田园

南宁

亮点： 明仕田园骑行

·全天

早起的游侠我们领队将带领骑行/徒步游览**明仕田园**

沿着村间小道，沿途山青水绿，村庄里的村民淳朴善良，没有商业化，没有喧闹的游客大军，一切都是那么的平静舒服。您也可以选择自费乘坐景区的游览接驳车在明仕田园逛逛，游览附近民风朴素的小村庄。或自行前往景区内乘坐竹筏，或自行闲逛于山水之间。酒店山水环绕，静静的呆在酒店看着青山绿水也是一件非常享受的事情。如不想早起的游侠可自由活动，睡到自然醒。

退房后品尝当地的农家菜午餐，适时集合返回南宁。

先送南宁机场（约15:30-16:00到达）**（飞机返程，请订18:00分以后的机票，如订太早的机票请自行安排前往机场）**

再送南宁东站（约16:30-17:00到达）**（高铁返程，请订17:30以后的高铁票，如订太早的高铁票请自行安排前往高铁站）**

不送南宁市区，可从南宁机场乘坐机场快线或从南宁火车东站乘坐地铁前往南宁市区。

车程：明仕山庄-南宁 约3小时 用餐：含早，正餐自理

备注：骑行明仕田园需视酒店当天的单车情况，如自行车不足或损坏较多，则改为徒步游览。  
骑行明仕田园，凭房卡可租自行车（会有比较老旧的情况），或可自费租用山地单车或电动摩托车等。因有一段公路骑行，老人家与小朋友不建议参与到骑行中。  
明仕山庄酒店泳池政策：需购买168元/张的游泳券方可进入泳池（具体以酒店前台为准）  
想继续停留在明仕之后前往其他目的地，可以乘坐10：30或16：30明仕-崇左南站的班车  

解散日：请预定南宁机场晚上18:00后的机票；请预定南宁火车东站晚上17:30后的高铁票； 不按页面时间订票，时间订早请自行安排返程交通

附加行程
----
 

费用说明
----

费用包含
----

成人

交通

D2-D4包车费、司机食宿津贴（4-5人7座商务车，6-13人安排19座中巴，14人以上安排常规旅游大巴出行）

住宿

南宁市酒店标间1晚（参考南宁宜尚Plus酒店或雅斯特国际酒店或携程4钻酒店)

靖西市酒店标间1晚（参考靖西雅斯特酒店或携程3钻酒店）

大新县明仕山庄标间1晚（可自费升级至明仕别院房型）

餐食

全程3个酒店早餐

门票及景交

德天跨国大瀑布首道景区门票

通灵大峡谷门票及景交车

鹅泉门票

渠洋湖游船门票

随团服务人员

随团人员服务（4-5人司机兼工作人员；6-15人安排1名游侠客服务人员；16-40人安排2名游侠客服务人员）

优惠补充说明

如持有军官证、伤残证、学生证、教师证等特殊证件，请在下单时备注，并携带原件，在团上向服务人员出示。  
因旅行社在景区门票上面是折扣票，优惠减免是按旅行社折扣价退还，不以景区挂牌价退还，优惠金额以景区政策为准。活动结束后线上退款/现场退款

儿童

交通

标准同成人

随团服务人员

标准同成人

其他

儿童费用不占床位（与成人共用床铺，若1成人单独携带儿童出行，请补1份房差安排1间房）、不含门票（可自行购买或由服务人员代买，1.2米~1.4米补景区门票约199元，1.4米以上儿童按成人门票）、不含早餐

自理费用
----

交通

各地往返南宁市的大交通费用（请在报名缴费成功后再预订大交通）

保险

建议您购买旅游意外险，游侠客提供不同的保险套餐，可在下单过程中自行选购

餐食

全程正餐，游侠客提倡自由用餐，根据自己口味选择，可约伴一起AA

门票及景交

景区里的二道门票或消费项目（如德天界河竹筏、德天高空滑道+飞天魔毯，自行选择）

单房差

行程中所列住宿标间含2个床位。单人报名的，游侠客担保拼房，根据报名顺序安排同性客人或领队、导游拼房；若不同意游侠客安排拼房，请补1份单房差以享用整个房间（标注多人间的除外）

赠送项目
----

赠送项目

游侠客定制头巾、精美田园靖西徽章、每人每天一瓶水

适合年龄
----

出行人

5 - 70岁

补充

报名咨询流程
------

在线报名

**活动页面“选择批次”，然后点击“立即报名”进行网上报名，按照要求提交有效信息（推荐）**

电话咨询

400-670-6300 \[周边游转“1”，国内游转“2”，出境游转“3”，团队定制转“4”\]

线路咨询

阿K17724508868,冬冬15158044986（ 咨询时间:工作日9:00-20:00）

企微咨询

打开游侠客APP或微信小程序，点击下方【进群】按钮，扫码入微信群咨询

微信公众号

**youxiake2008**

活动说明
----

集合时间：批次第一天南宁全天集合。

集合地点：南宁市区指定酒店

成行人数：缴费满4人成行，4-5人安排小车司机接待（不陪同进园，客服会联系确认）；6人及以上安排中巴+游侠客领队带队；16人以上安排双领队带队；每车人数不超过40人；

成团须知：欢迎提前报名，报名人数达成行人数，客服会通知成行缴费，一般提前7~12天；若订单报名状态为“已通过”，代表此批次已经成行，即可支付活动费用，预定大交通；如因游侠客原因不能成行，游侠客将承担已缴费队员机票退票损失（需提供官方损失凭证）。

报名活动流程
======

查看活动 注册报名 审核通过 付款 快乐出游 发游记

预订须知
----

查看更多∨

安全提示
----

查看更多∨

注意事项
----

1.若遇恶劣气候影响容易造成道路阻断，行车时间只能作为参考。

2.鹅泉、德天瀑布、通灵大峡谷等观水景点，水深危险，携带小孩的客人请多加注意。

3.大家有不同的出游习惯，但团队活动需听从领队安排，遇到突发情况请领队联系协商解决。

4.注意安全，注重环保，尊重当地百姓，发扬游侠互助精神，友善对待遇到的每个人、每件事。

5.若遇人力不可抗拒情况，无法继续既定行程，领队可根据实际情况有权更改或调整行程。

6.若因自身原因，中途退出活动，门票费用退还，房车费用照计，请理解。

7.参加本次活动者视为遵守以上出团约定。

查看更多∨

活动装备
----

退订政策
----

查看更多∨

特殊原因退订保障
--------

为了分担游侠客旅游客户因为自身不能预期和控制的特殊原因退订产生的损失，游侠客特别设立了“特殊原因退订保障”。

  

若您的情况在下方特殊原因保障范围内，请您在申请退订时选择原因为“特殊退订”，并在提出申请后七天内提交下方要求的必备申请材料至订单的退订详情中，您的情况与材料经游侠客审核确认完全符合“特殊原因退订保障政策要求”后，游侠客将补贴给您：订单最终实际损失的50%金额的游侠客“特殊退订优惠券”（已生效的保险费用及已送签并已产生的签证可选项费用不算入订单实际损失）。

  

1、游侠客“特殊退订优惠券”使用规则：

**有效期：一年，此优惠券无法延期，请您务必关注好您帐户下优惠券有效期限，以免过期。**

**使用范围：游侠客平台内跟团游、半自助游、邮轮游、自由行品类下的产品订单。**

满减标准：

|     |     |
| --- | --- |
| 面额  | 满减金额 |
| 金额≤50 | 单笔旅游费用总额满200 |
| 50＜金额≤100 | 单笔旅游费用总额满400 |
| 100＜金额≤200 | 单笔旅游费用总额满800 |
| 200＜金额≤300 | 单笔旅游费用总额满1200 |
| 300＜金额≤500 | 单笔旅游费用总额满2000 |
| 500＜金额≤800 | 单笔旅游费用总额满3200 |
| 800＜金额≤1000 | 单笔旅游费用总额满4000 |
| 1000元以上将按照以上面额拆分；不可与其他优惠券同时使用。 |     |

2、特殊原因退订保障政策要求：

|     |     |     |     |
| --- | --- | --- | --- |
| 状况当事人 | 突发情况 | 特殊原因退订受益人 | 材料及说明 |
| 出行人本人 | 航班延误 | 出行人本人 | 航班延误证明、登机牌、身份证正反面（出境线路提供护照信息页） |
| 出行人本人及其子女、配偶、父母（含岳父母、公婆） | 身故  | 订单内本人及同行子女、配偶、父母（含岳父母、公婆） | 死亡证明、户口本等关系证明材料。 |
| 住院  | 出具三甲医院病例/诊断书，包含“不宜出行”字样的病假单、住院证明（住院期间须与出行期间重叠）和收费证明。<br><br>特别说明：<br><br>1、仅保障因突发急性病的住院。（【突发急性病】指在本次产品预订后出行日期前突然发生的、在突发前未曾接受治疗或诊断的（首次罹患的突发性疾病或出现的）、必须立即接受治疗方能避免身体或生命伤害的疾病，不包括既往疾病、慢性病、精神病、 精神分裂、艾滋病、性传播疾病、遗传性疾病、先天性疾病或缺陷、先天性畸形、牙齿治疗、预防性手术、非必须紧急治疗的手术或器官移植、因脊椎间盘突出症或错位的治疗、脑梗、心梗等。）<br><br>2、若病历/诊断书/病假单上证明病情/怀孕日期在产品预订日期之前则不做保障。<br><br>3、当地若无三级甲等医院的以当地最高级别医院诊断为准。 |
| 骨折  |
| 出行人怀孕 | 怀孕  | 本人及同行子女、配偶 |
| 出行人本人 | 拒签  | 拒签本人 | 拒签需出具使领馆拒签信（出具日期在出行日期前），以下情况不再保障之列：<br><br>1、单签证产品。（若在预订签证产品同时预订同签字目的地的团队、半自助游、邮轮产品的，则可以整体享受保障。）<br><br>2、签证材料整理服务等需游客自行送签产品（此类产品游侠客未收取领馆签证费和领馆签证中心服务费）。<br><br>3、您不能按照领馆要求或追加要求的材料清单提供充分真实的材料，或者您在该国家/地区已经存在不良记录。<br><br>4、您被拒签日期在旅游产品预订日期之前。<br><br>5、因领馆系统观音导致的签证延迟出签。 |
| 特别说明：<br><br>1、旅游者购买旅游服务前，需重点阅读旅游产品的退订政策、预订须知、安全须知，并结合自身身体状况做出是否预订，是否适合出行的评估。<br><br>2、逾期未联系游侠客申请并按要求递交保障所需材料的视为自动放弃，游侠客有权不予保障。<br><br>3、特殊原因退订所需申请材料当事人需保证材料真实性，若材料存在真实性问题，游侠客有权不予保障，并有权收回已发放的保障。<br><br>4、受理条件以受理当日公示的受理条件为准，游侠客保留在实当时候修订受理条件而不另外通知的权利。 |     |     |     |

查看更多∨

价格说明
----

**·划线价：**

指商品的供应商指导价、门市价或曾经该商品展示过的销售价等，并非原价，仅供参考。

**·未划线价格：**

商品在游侠客平台的实时售价，不因表述的差异改变性质。具体成交价格根据用户使用的优惠券、游侠币抵扣、游侠客平台余额或商品参加特定优惠活动和时段波动等情形发生变化，最终以订单结算页价格为准。

查看更多∨

常见问题
----

1.如何前往市区酒店宜尚Plus店。  
机场 ： 网约车大概80元 打车130左右 或者乘机场大巴1号线至民航酒店下车，再打车至酒店约8元左右或者步行1公里,还可以查吴圩机场到南宁站的动车，时间合适可以16分钟到南宁站，出来打车或者步行到酒店.

南宁东站 ： 打车约45元左右，也可以地铁1号线到南宁火车站步行1公里或者打车8元

南宁火车站 : 步行1公里或者打车8元

查看更多∨

用户评价_（2177）_
------------

*   全部 (2177)
*   最新
*   当季 (572)
*   有图 (1164)
*   优质 (38)

*   热情洋溢(142)
*   满分服务(132)
*   朋友式的旅行(128)
*   鉴定为优秀领队(110)
*   知识渊博(105)
*   嗨到不行(85)
*   会推荐朋友参加(74)
*   行程物超所值(71)

_查看更多∨_

![游侠客旅行](http://img2.youxiake.com/avatar12827420_juzqTiDnfbB8.jpg!x200) 青\*\*\*

**行程**_非常满意_**领队**_非常满意_

不来广西，怎么知道这里是真正的不夜城夜市开到天亮。不来德天大瀑布，怎么感受近距离的瀑布震撼。短短4天的广西之行，在领队小健的带领下圆满的结束了。领队非常好，会介绍南宁各种好吃的，我们拍照脱队了也会等我们。 在短短的4天从南宁-靖西-德天，广西山水秀美给我们留下了深刻的印象。第一天在南宁集合酒店很好而且房间都是智能化，在领队的推荐下吃了桂小厨，老友鱼酸酸微辣越吃越香，黑豆粉那是一个鲜，百香果拌黄瓜打开新世界，吃完老友粉晚上20点要去邕江码头看打铁花，23点以后去夜市感受南宁越夜越嗨。第二天8点出发去通灵大峡谷，领队一路介绍，到了峡谷还会介绍当地植物，幸运的是瀑布水量很大，但水量大就无法穿过水帘洞看溶洞，爬了阶梯做了有氧，渠洋湖纯休闲坐船欣赏风景，这里提个小建议其实如果是淡季没必要8点出发，如果是9点出发会更舒服。 晚上住在靖西，我推荐去龙潭湿地公园看看当地人的生活吃个番叽嘟，旁边两个商场想吃什么吃什么，领队推荐的螺蛳粉也很好吃，20点可以去锦绣古镇看篝火，不过这个古镇没啥好吃好玩的。第三天重头戏德天大瀑布了非常值得，坐船近距离观赏幸运瀑布而且水量很大非常壮观，晚上住就是明仕田园可以看个篝火表演，当地吃的我推荐勇哥烧烤，爆炒石螺和炒米粉非常好吃，当地的漓泉啤酒也值得一尝。第四天早上还可以骑行，不过因为带着我妈她不会骑车我们只能花钱做观光车了感受田园风光也是很惬意的。整个行程紧凑也合理值得一游。

_展开全部∨_
 
2024-05-14 22:46:48 **此评价来自** _游侠客APP_

![游侠客旅行](http://img2.youxiake.com/001/26/67/04_avatar_big.jpg!x200) 2\*\*\*

**行程**_非常满意_**领队**_非常满意_

💖旅途感受： 有一个会吃会玩会拍照的领队是怎么样的体验，那一定是拥有Tina的快乐，不得不说很高兴遇见这么优秀领队，一路给大家安排的明明白白，妥妥当当的。沿途的风景也是很美丽，很早就想来靖西这条路了，一直没有凑到机会，这次终于梦想成真，收获了不一样的美好。 👍推荐理由： 通灵大峡谷简直太动植物百科全书了，分分钟走进绿叶仙踪，放下手机哐哐吸氧。德天大瀑布是最想去的一个地方，一面是越南，一面是中国经常还有小贩在河上兜售物品，虽然这个季节水量没有那么丰富，但好天气下的绿水青山拍起来太美了，不恐高的小伙伴一定要体验一下，玻璃栈道上俯瞰美景以及高速滑滑梯。 🌷行程亮点： 名仕是下雨最大的一天，简直是一场雨把我困在山水画里，收官之日主打一个推背感，本来想着是休闲田园骑行，秒变大雨灌脸＋乡野迷踪，电瓶车铆劲上坡刹车下坡，被大雨浇得路都看不见了，只能借村户家铁棚躲躲雨，还好有一群疯批骑友一起一路欢腾。 这一路上经历了不一样的风景，不一样的天气，不一样的来自五湖四海的小伙伴，还有游侠客优秀的领队Tina，避开人群的游玩，实在太棒了，满满的收获，美景太多了，快乐太多了，无法言语，不如多看看图片吧。
 
2024-05-06 23:56:36 **此评价来自** _游侠客APP_

![游侠客旅行](http://img2.youxiake.com/mini/SYjnWnLjd8RUcQRFv6cX.jpg!x200) 段\*\*\*

**行程**_非常满意_**领队（1）**_非常满意_**领队（2）**_非常满意_

五一放假前刷🍠看到崇左的绝美风景 本P人说走就走 马上集结朋友们做攻略 本想自由行 奈何景点太远租车麻烦 又定不上酒店 赶紧上游侠客看有没有合适的行程 天时地利人和 这条线路和我们的时间和预想要去的景点高度重合 马上报名 果然没让人失望 广西的风景真的好好 好山好水好风光 太适合都市隶人出去放松了 大自然治愈一切 除了我的腿😭 领队小贱 很细心 讲解充分 了解到了很多广西地域特色以及民俗 每个景点的安排都很好 尽可能避开大人流 这几天游玩需要排队的时间都比较少 这点很满意 领队立夏 美女小姐姐一枚 温柔耐心 一直帮大家拍照 果然没带单反是正确的选择 嘿嘿嘿 立夏辛苦啦 领队帅照美照 附图9.图10 总体来说这次也是很满足的一次旅行！期待下次出发！

_展开全部∨_
 
已经是第一张了

2024-05-06 21:34:55 **此评价来自** _游侠客APP_

_客服 回复：_ 亲爱的游侠，当我们的产品和服务得到您的肯定时，是我们最开心的时刻。期待与您的下次相聚！

![游侠客旅行](https://img2.youxiake.com/2020/10/29/avatar/shiziwang_3x.png!x200) 游\*\*\*

**行程**_非常满意_**领队（1）**_非常满意_**领队（2）**_非常满意_

这次旅行非常完美。感谢领队小美、小陆及司机师傅的一路帮助和陪伴！首先行程安排合理，时间充足。其次天气适宜，永远是路上倾盆大雨，而到达景点后雨就停了。 D1通天大峡谷：为了看落差188米的瀑布，穿越“通天通地”的大峡谷，返程有一段带有典型喀斯特地貌特点的溶洞，持续滴水的小路通往洞穴深处，洞穴内石笋石芽交替排列，又有小溪流过，不禁感叹大自然的鬼斧神工，此洞很适合拍摄仙侠剧\[偷笑\]一路上各种原始植被，还有各色蝴蝶蜻蜓在四周盘旋，头顶再时不时来一张被大型彩色蜘蛛占据着的网，难得一见的亚热带丛林风光\[胜利\] 渠洋湖：没看到彩虹夕阳有点可惜，但平静的湖面上一叶孤舟划过，仍在这暑期的旅游旺季给人带来了内心的静谧 D2鹅泉：一个村庄因为一位老奶奶救了两只小鹅，留下泉眼的传说而得名。景色很美，在湖边休息时，看着游人如织，不禁在想如果美景中没有人，又会怎样？ 德天跨国瀑布：5A景区，亚洲第一、世界第四大跨国瀑布，瀑布并不高，但水流喘急，宛如银河倾泻。从高处俯瞰，多角度解锁“一瀑两国”。对我而言，比较有趣的是第一次坐魔毯，然后过一段玻璃栈道后再坐“滑滑梯”下来，在山与水之间，仿佛又回到了儿时的快乐时光 D3明仕田园：据说这里是花千骨的取景地。早上可以睡到自然醒，也可沿着村间小道，在青山绿水间骑行。
 
2023-08-27 20:01:08 **此评价来自** _游侠客APP_

![游侠客旅行](http://img2.youxiake.com/default_avatar.png!x200) 欧\*\*\*

**行程**_非常满意_**领队（1）**_非常满意_**领队（2）**_非常满意_

非常完美的一次旅行，首先是天气，预报南宁天天下雨，第一天一早从南宁出发大雨，结果开到景区天晴了，后面几天几乎都如此，在大巴上下雨，到了景区都只剩蒙蒙细雨啦，绝对好人品，哈哈。我的旅游运一直不错，毕竟去峨眉山也是看到了金顶，去土耳其只放了一天热气球就飞成功了。 再来说说景点，每一天都比前一天更好，第一天的灵通大瀑布说实话比较一般，主要是挺累的，要往下走楼梯，导致后面几天小腿肚都很酸，腿脚不好年纪大的建议可以舍弃这个景点，下午的游湖很舒服。 第二天的景点类似小家碧玉的精致园林，有山有水有倒影，还是挺不错的。第三天才是重点，也就是德天大瀑布，一定要自费去坐魔毯上到最高观景台，那里视野开阔，类似无人机视角，可以一览众山小，还可以看到对面的越南村庄，绝对超值，最后还可以滑滑梯滑下来，还是蛮有趣的。 最后一天的明仕山庄，田园风情还不错，只是没有太惊艳的景点，推荐有空还是可以自费购买游泳票去游个泳，泳池还是挺漂亮的，会让你觉得更加不虚此行！非常喜欢这个有点小众的线路，行程安排不错，景色也大为惊喜！ 最后感谢导游小美和咕咕陆，服务周到细心，还有司机师傅开车很稳很不错！
 
尾页

活动咨询
----

![](https://www.youxiake.com/old_yxkweb/Public/assets/img/common/avatar50x50.png)

发布

![游侠客旅行](https://img2.youxiake.com/avatar11515122_j0VXMB2c4luJ.jpg!x200)

#### [小刚](https://www.youxiake.com/u/11515122)
:

我想带我母亲一起参加，她78岁，可以吗？

2024-06-18 18:32:08

评论

_游小侠_ 回复了 [小刚](https://www.youxiake.com/u/11515122)
 您好，非常抱歉，这个年龄超过我们的报名年龄范围

**2024-06-19 10:02:38**

评论

![游侠客旅行](https://img2.youxiake.com/avatar11858985_KRp1emAI0afU.png!x200)

#### [我是一颗大树🌳](https://www.youxiake.com/u/11858985)
:

端午都是下大雨，这还能出游吗？？？

2024-06-03 18:07:46

评论

_游小侠_ 回复了 [我是一颗大树🌳](https://www.youxiake.com/u/11858985)
 您好，目前当地的天气预报以为阵雨小雨为主，山区天气变化比较快。路线涉及的都是成熟的景区步行道，欢迎报名参加。

**2024-06-04 10:53:26**

评论

![游侠客旅行](https://img2.youxiake.com/avatar10663593_CwcaZ0C900yV.jpg!x200)

#### [风景格](https://www.youxiake.com/u/10663593)
:

下雨影响游玩体验吗？

2024-05-07 14:26:51

评论

_游小侠_ 回复了 [风景格](https://www.youxiake.com/u/10663593)
 您好，现场会根据实际天气情况来调整游玩顺序或稍等雨势变小，大部分景点都是在景区里面，有对应的游客休息区，现场都是安全的。

**2024-05-15 09:23:41**

评论

![游侠客旅行](https://www.youxiake.com/images/no_avatar.png)

#### [雪梨町](https://www.youxiake.com/u/1200420)
:

4.28有女生同行吗

2024-04-25 12:40:18

评论

_游小侠_ 回复了 [雪梨町](https://www.youxiake.com/u/1200420)
 您好，您自己一个人也可以报名，我们会安排同性游侠一起拼房，不需要额外承担单房差

**2024-04-28 11:19:38**

评论

![游侠客旅行](https://img2.youxiake.com/avatar10526339_pwAADttG0Z09.jpg!x200)

#### [风铃草](https://www.youxiake.com/u/10526339)
:

集合日酒店几点可以入住

2024-04-23 23:35:39

评论

_游小侠_ 回复了 [风铃草](https://www.youxiake.com/u/10526339)
 您好，集合酒店一般情况需要在下午2点后入住，提前到达可以在前台寄存行李。

**2024-04-28 11:20:25**

评论

首页

上一页

1234567891011121314151617181920212223242526272829303132333435363738394041424344454647484950515253545556575859606162636465666768697071727374757677787980

下一页

尾页

目的地介绍
-----

1.德天大瀑布：位于中越边境广西大新县硕龙镇，为国家特级景点。横跨中国越南两个国家，是世界第四大、亚洲第一大跨国瀑布.

2.明仕田园：属于大新县堪圩乡明仕村。明仕田园风光，距县城53公里，为国家一级景点。方圆20公里的景区山清水秀，山环水绕，素有小桂林之称。这里翠竹绕岸，农舍点缀，独木桥横，稻穗摇曳，农夫荷锄，牧童戏水，风光俊朗清逸，极富南国田园气息。美丽的山峰，明镜似的小河，百鸟啼鸣，以及村舍中的鸡鸣狗吠，是一处难求的令人迷醉的田园风光。

3.靖西：历史悠久，典型的壮族人口聚居地，年均气温19.1℃，素有气候“小昆明”之称。境内以溶蚀高原地貌为主，山明、水秀，以奇峰异洞、四季如春的自然风光闻名遐迩，又有山水“小桂林”之誉，是旅游、度假和避暑的理想胜地。

查看更多∨

### 图文介绍

通灵大峡谷是亚洲最高的单极瀑布，于靖西县城东南部32公里处，全长2.8公里，谷全长2.8公里，由通灵峡、念八峡及地下暗河、隧道贯通连接，通天彻地，灵气飘逸。

![田园靖西 | 广西崇左田园风光，德天跨国瀑布4日](https://gallery.youxiake.com/product/line/202308/11/1691721615397.jpg)

渠洋湖是一个人工湖，因为地处渠洋镇而称之为渠洋湖，湖中间，十几座形状各异、高矮不一、怪石嶙峋的独峰拔水而起，水面上的山错落突兀、争奇竟秀、风景优美，而被人们誉为“水上桂林”。 傍晚时的渠洋湖，水面波光粼粼，夕阳的余晖照在湖面的小舟上。而远处的山峦如黛，在绯红的霞光中渐渐隐去，天空中偶有雁群飞过，再美的丹青水墨画也不过如此。

![田园靖西 | 广西崇左田园风光，德天跨国瀑布4日](https://gallery.youxiake.com/product/line/202405/06/1714985337860.jpg)

鹅泉距离靖西县城4.5公里，与云南大理蝴蝶泉、广西桂平西山乳泉并称为中国西南三大名泉。靖西有八景，其中一景“鹅泉跃鲤三层浪”，说的便是鹅泉中鲤鱼听人呼喊便跃出水面的奇观。

![田园靖西 | 广西崇左田园风光，德天跨国瀑布4日](https://gallery.youxiake.com/product/line/202302/28/1677574330913.jpg)

鹅泉观景台

![田园靖西 | 广西崇左田园风光，德天跨国瀑布4日](https://gallery.youxiake.com/product/line/202405/06/1714985480503.jpg)

德天跨国瀑布，地处归春河上游，横跨中越两国，以层叠秀美著称，是亚洲第一的跨国瀑布。三级落差的德天瀑布与越南的板约瀑布隔河相望。

![田园靖西 | 广西崇左田园风光，德天跨国瀑布4日](https://gallery.youxiake.com/product/line/202308/11/1691721784172.jpg)

德天瀑布观景台

![田园靖西 | 广西崇左田园风光，德天跨国瀑布4日](https://gallery.youxiake.com/product/line/202405/22/1716368829791.jpg)

明仕田园风光，这里山清水秀，山环水绕，素有“小桂林”之称，是电视剧《花千骨》的取景地。

![田园靖西 | 广西崇左田园风光，德天跨国瀑布4日](https://gallery.youxiake.com/product/line/202305/09/1683596006776.jpg)

明仕田园风光

![田园靖西 | 广西崇左田园风光，德天跨国瀑布4日](https://gallery.youxiake.com/product/line/202405/06/1714985337875.jpg)

[](https://www.youxiake.com/dingzhi)

报名情况 [更多](https://www.youxiake.com/allorders.html?pid=22510)
 
最终的结果：markdown语法的数据应该是如下形式:
# 产品信息摘要

## 产品名称
田园靖西 | 广西崇左田园风光，德天跨国瀑布4日

## 价格
- 成人：￥1780元起
- 儿童：￥950元起

## 游玩的天数
4天3晚

## 描述
南宁集散 | 探索中国侏罗纪公园-通灵大峡谷 | 深入边陲西南小城-靖西 | 确保入住明仕景区内的明仕山庄

## 产品特色
- 小众目的地
- 西南秘境
- 明仕篝火晚会
- 峡谷避暑

## 出发地
南宁

## 行程套餐的名称
入住明仕山庄酒店

## 批次套餐的名称
南宁集散

## 标签
跟团游 · 生态乡野 · 小众风光 · 游山玩水

## 每个批次日期对应的价格
- 2024-07-14(周日)-2024-07-17(周三) 成人:1980元 儿童:1050元
- 2024-07-15(周一)-2024-07-18(周四) 成人:1980元 儿童:1050元
- 2024-07-16(周二)-2024-07-19(周五) 成人:1980元 儿童:1050元
- 2024-07-17(周三)-2024-07-20(周六) 成人:1980元 儿童:1050元
- 2024-07-19(周五)-2024-07-22(周一) 成人:1980元 儿童:1050元
- 2024-09-10(周二)-2024-09-13(周五) 成人:1780元 儿童:950元
- 2024-09-22(周日)-2024-09-25(周三) 成人:1780元 儿童:950元

## 活动亮点
- 花样好评：年收客2000+人，每团均花式好评，多年独立操作，放心报名
- 小众目的地：广西西南边境，游览西南边境风景极秀丽的景观
- 通灵大峡谷：走进清凉峡谷，落差188.6米的西南第一高瀑布
- 德天跨国瀑布：亚洲第一、世界第四大跨国瀑布
- 渠洋湖游船：被称为山中的大海，泛舟在“广西版下龙湾”景色中赏日落时刻
- 西南名泉鹅泉：西南地区三大名泉美称的鹅泉
- 入住明仕山庄：特别安排入住世外桃源明仕山庄
- 深度纯玩：全程纯玩无购物，游侠客领队带队朋友式旅行

## 每日的具体行程
- D1：南宁集合日
- D2：南宁至通灵大峡谷、渠洋湖，后入住靖西市区酒店
- D3：靖西至鹅泉、德天大瀑布，后前往明仕山庄
- D4：明仕山庄至明仕田园，最后返回南宁

### D1的行程安排
·全天
全天南宁集合日。自行前往入住游侠客指定酒店。
集合酒店位于市区，交通便利，早到的游侠可去周边游玩和寻觅美食，自由活动期间请妥善保管随身携带物品，注意安全！
**自由活动推荐**
南湖、青秀山、民族博物馆、三街两巷、南宁国际会展中心、航洋国际、南湖公园、青秀山景区等。
前往美食街区中山路，建政路夜市，当然少不了品尝当地的美食，如老友粉、手榨米粉、糖水等等。
**集合酒店交通指引**
南宁吴圩机场 ： 网约车约90元~120元， 或者乘机场大巴1号线至民航酒店下车，再打车至酒店（约10元）
南宁东站： 打车约50元，也可以坐地铁1号线到南宁火车站步行到酒店
南宁火车站：出站后步行1公里或者打车（约10元）
用餐：全天自理
住宿：南宁酒店标准双人间（参考南宁宜尚Plus酒店或雅斯特国际酒店或网评同级酒店）
备注：一般提前2~3天会建群联系

### D2的行程安排
亮点： 打卡西南瀑布通灵大峡谷 渠洋湖游船赏日落
·全天
早餐后乘坐大巴前往通灵大峡谷。
车上相互自我介绍，领队介绍景区，并做一些有趣的互动小游戏，在欢声笑语中迎接接下来的旅途。
到达景区门口后我们先AA腐败解决午餐，随后跟领队一起进入景区。沿着景区的楼梯向下穿越溶洞，来到峡谷底，峡谷内水流潺潺，拥有着举世罕见的西南第一高瀑布—通灵大瀑布，同时峡谷内还拥有洞中瀑布、地下暗河、洞穴奇观、古石垒、古悬葬、原始植被等自然景观与人文景观。
适时集合，我们驱车前往渠洋湖，这里是一个喀斯特高原湖泊水库，峰林小岛星星点点散落在湖中，配合上清澈透蓝的湖水，拍个照片发到朋友圈，别人还以为你在下龙湾呢。将近日落时分，湖水倒影着夕阳，我们乘坐游船慢慢欣赏，惬意至极。
游览完毕后乘车入住靖西市区酒店。
车程：南宁-通灵大峡谷约3小时；通灵大峡谷-渠羊湖1.5小时；渠羊湖-靖西酒店约50分钟 用餐：含早，正餐自理，可AA腐败
住宿：靖西酒店标准双人间（参考：靖西雅斯特酒店或同级酒店）
备注：假期日期间，通灵大峡谷可能会有堵车的情况，如影响渠羊湖看日落的时间，领队有权调整行程为鹅泉景区，渠羊湖改为第二天的上午行程。 游玩通灵大峡谷景区需要上下景区楼梯，下楼梯路段大概818级楼梯，上楼梯的路段相对平缓点，没有严重的膝盖问题均可参加。

### D3的行程安排
亮点： 名泉靖西鹅泉 德天跨国大瀑布
·全天
早餐后集合后驱车前往鹅泉
这里是一处面积达到50平方公里的巨泉，有西南三大名泉的美称，泉水由透明到绿色到蓝色，泉眼位置的水深达到了30米，德天瀑布的源头就是此处，我们可以在泉水边静静的发呆，也可以攀登到半山亭俯瞰鹅泉全景。
午餐后接着乘坐大巴前往德天跨国大瀑布
德天大瀑布，取得天独厚之意，一半属中国、一半属越南，一不小心就出国了。激流从70米高山崖飞泻而下，三层跌落如万马奔腾，轰鸣如水龙吟，《花千骨》长流山仙界的取景地。（节假日期间德天景区乘坐景区接驳车的人比较多，遇到排队的情况，请跟随团队耐心等候）
之后前往明仕田园
到达明仕田园后，我们将入住位于景区内的明仕山庄酒店，酒店被山水包围、诗情画意、颇有隐居仙境之意。游侠们可在此赏景、喝茶、漫步、发呆、闲聊。
晚上我们围着篝火晚会一起跳舞，和来自各地的游侠们一起感受热闹的夜晚。

## 费用说明
- 费用包含：交通、住宿、餐食（全程3个酒店早餐）、门票及景交、随团服务人员等。
- 儿童费用：不占床位、不含门票、不含早餐。
- 自理费用：各地往返南宁市的大交通费用、保险、全程正餐、景区里的二道门票或消费项目、单房差等。
- 赠送项目：游侠客定制头巾、精美田园靖西徽章、每人每天一瓶水

## 活动说明
集合时间：批次第一天南宁全天集合。
集合地点：南宁市区指定酒店
成行人数：缴费满4人成行，4-5人安排小车司机接待（不陪同进园，客服会联系确认）；6人及以上安排中巴+游侠客领队带队；16人以上安排双领队带队；每车人数不超过40人；

## 常见问题
1.如何前往市区酒店宜尚Plus店。
机场 ： 网约车大概80元 打车130左右 或者乘机场大巴1号线至民航酒店下车，再打车至酒店约8元左右或者步行1公里,还可以查吴圩机场到南宁站的动车，时间合适可以16分钟到南宁站，出来打车或者步行到酒店.
南宁东站 ： 打车约45元左右，也可以地铁1号线到南宁火车站步行1公里或者打车8元
南宁火车站 : 步行1公里或者打车8元

## 适合年龄
5 - 70岁

## 目的地介绍
- 德天大瀑布：位于中越边境广西大新县硕龙镇，为国家特级景点。
- 明仕田园：属于大新县堪圩乡明仕村，为国家一级景点。
- 靖西：历史悠久，典型的壮族人口聚居地，有山水“小桂林”之誉。

Extracted:"""
```
