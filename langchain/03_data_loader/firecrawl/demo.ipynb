```python
from dotenv import load_dotenv
from langchain_community.document_loaders import FireCrawlLoader
import os

load_dotenv()
firecrawl_api_key = os.getenv("FIRECRAWL_API_KEY")
```

```python
# os.environ["FIRECRAWL_API_URL"] = 'https://crab233.cloudns.biz/proxy/api.firecrawl.dev'
os.environ["FIRECRAWL_API_URL"] = 'https://api.firecrawl.dev'
```

```python
scrape_url = 'https://www.youxiake.com/lines.html?id=47233&spm=eyJmcm9tIjoxMDgwLCJvcmlnaW5hbF9pZCI6MH0'
loader = FireCrawlLoader(api_key=firecrawl_api_key, url=scrape_url, mode="scrape")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

documents = loader.load()
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
docs = text_splitter.split_documents(documents)
```

```python
docs
```

```python
def format_docs(docs):
    context = "\n".join(doc.page_content for doc in docs)
    return context
```

```python
context = format_docs(documents)
context
```
