```python
from langchain_community.document_loaders import UnstructuredURLLoader
from langchain_community.document_loaders import WebBaseLoader
```

```python
urls = [
    "https://www.youxiake.com/lines.html?id=26368&spm=eyJmcm9tIjo0MTMsIm9yaWdpbmFsX2lkIjoyNjM3M30",
]
```

```python
loader = UnstructuredURLLoader(urls=urls)
docs_UnstructuredURLLoader = loader.load()
docs_UnstructuredURLLoader
```

```python
loader = WebBaseLoader(urls)
# loader.requests_kwargs = {'verify': False}
docs_WebBaseLoader = loader.load()
docs_WebBaseLoader
```

```python
from langchain_community.document_loaders import ToMarkdownLoader
import os
from dotenv import load_dotenv

load_dotenv(override=True)

markdown_key = os.environ.get("2MARKDOWN_KEY")
print(markdown_key)

loader = ToMarkdownLoader(url=urls[0], api_key=markdown_key)
docs_ToMarkdownLoader = loader.load()
docs_ToMarkdownLoader
```

```python
print(docs_ToMarkdownLoader[0].page_content)
```

```python
from langchain_community.document_loaders import (
    DirectoryLoader,
    UnstructuredMarkdownLoader,
)

loader = DirectoryLoader(
    ".", glob="**/*.md", loader_cls=UnstructuredMarkdownLoader
)
documents = loader.load()
```
