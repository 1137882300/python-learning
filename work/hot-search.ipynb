```python
import requests
from bs4 import BeautifulSoup


# Example function for Baidu search
def get_baidu_search_hotness(keyword):
    url = f'https://www.baidu.com/s?wd={keyword}'
    headers = {'User-Agent': 'Mozilla/5.0'}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract specific search hotness information
    # This part is highly dependent on the page structure and might need adjustments
    hotness_info = soup.find('div', {'class': 'result'}).text

    return hotness_info


# Example function for 360 search
def get_360_search_hotness(keyword):
    url = f'https://www.so.com/s?q={keyword}'
    headers = {'User-Agent': 'Mozilla/5.0'}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    hotness_info = soup.find('div', {'class': 'res-list'}).text

    return hotness_info


# Example function for Bing search
def get_bing_search_hotness(keyword):
    url = f'https://www.bing.com/search?q={keyword}'
    headers = {'User-Agent': 'Mozilla/5.0'}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    hotness_info = soup.find('li', {'class': 'b_algo'}).text

    return hotness_info


# Function for other platforms (e.g., WeChat, Douyin, Xiaohongshu)
# Usually, these platforms don't provide direct search API, so you may need to use Selenium for automation

def get_search_hotness(keyword):
    platforms = {
        'Baidu': get_baidu_search_hotness,
        '360': get_360_search_hotness,
        'Bing': get_bing_search_hotness,
        # Add other platforms' functions here
    }

    hotness_data = {}
    for platform, func in platforms.items():
        try:
            hotness = func(keyword)
            hotness_data[platform] = hotness
        except Exception as e:
            hotness_data[platform] = f'Error: {str(e)}'

    return hotness_data


# Example of how to use the function
keywords = ['阿尔山', '东北秋色', '额济纳旗', '稻城亚丁', '雨崩', '北疆']

for keyword in keywords:
    print(f"Hotness data for '{keyword}':")
    hotness_data = get_search_hotness(keyword)
    for platform, hotness in hotness_data.items():
        print(f"{platform}: {hotness}")
    print()
```
