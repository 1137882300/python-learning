```python
import os
import pandas as pd
import requests

# 读取 Excel 文件
excel_file = '/Users/pangmengting/Downloads/文化场馆表.xlsx'
df = pd.read_excel(excel_file, engine='openpyxl')

# 创建保存图片的文件夹
# output_folder = 'images'
# os.makedirs(output_folder, exist_ok=True)

# 遍历每一行,下载图片并用标题命名
for index, row in df.iterrows():
    org_type = row['org_type']
    poi_id = row['poiID']
    category = row['分类名称']
    title = row['名称']
    address = row['地址']
    city = row['城市']
    county = row['区县']
    lo = row['经度']
    la = row['纬度']
    open_time = row['开放时间']
    pic = row['图片']
    if index == 100:
        break
    print(index, org_type, poi_id, category,title,address,city,county,lo,la,open_time, pic)
```
