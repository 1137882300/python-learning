```python
class VenueInfo:
    def __init__(self, org_type, category_id, category, title, address, city, county, lo, la, open_time, pic):
        self.org_type = org_type
        self.category_id = category_id
        self.category = category
        self.title = title
        self.address = address
        self.city = city
        self.county = county
        self.longitude = lo
        self.latitude = la
        self.open_time = open_time
        self.picture = pic

    def __str__(self):
        # 使用 __dict__ 将对象的属性转换为字典
        # 然后将字典转换为 JSON 字符串
        return json.dumps(self.__dict__, ensure_ascii=False, indent=4)

    def to_dict(self):
        # 转换为字典
        return self.__dict__
```

```python
import requests
import json

error_list = []


def process_picture(picture_url):
    print(picture_url)
    # API 请求的URL
    url = 'https://fxhz.wgly.hangzhou.gov.cn/dashboard/api/oss/transfer/2oss'

    # 请求的头部
    headers = {
        'User-Agent': 'Apifox/1.0.0 (https://apifox.com)',
        'Content-Type': 'application/json'
    }

    # 请求的正文
    payload = json.dumps({
        "urlList": [picture_url],
        "separator": ".cn/whyxoss/"
    })

    # 发送 POST 请求
    try:
        response = requests.post(url, headers=headers, data=payload)

        # 检查响应状态码
        if response.status_code == 200:
            # 假设返回的 JSON 中有一个 'data' 字段
            data = response.json().get('data')
            return json.loads(data)[0]
        else:
            error_list.append(picture_url)
            print(f"请求失败，状态码: {response.status_code}")
            return 'oooo'
    except Exception as e:
        error_list.append(picture_url)
        print(f"处理图片过程中发生异常: {e}")
        return 'oooo1'
```

```python
# 定义城市和区县的映射
city_county_map = {
    '杭州市': {
        '上城区': '330102',
        '拱墅区': '330105',
        '西湖区': '330106',
        '滨江区': '330108',
        '萧山区': '330109',
        '临平区': '311199',
        '钱塘区': '310020',
        '富阳区': '330183',
        '临安区': '330185',
        '建德市': '330182',
        '桐庐县': '330122',
        '淳安县': '330127',
        '余杭区': '330110',
        '西湖风景区': '4542427'
    }
}


def get_county_code(city, county):
    """根据城市和区县名称获取对应的代码"""
    city_map = city_county_map.get(city)
    if city_map:
        return city_map.get(county, '0')  # 默认返回杭州市的代码
    return '0'
```

```python
import re


def validate_and_default_longitude(longitude):
    # 匹配经度格式：±DD.DDDDDD
    pattern = r'^[+-]?\d{1,3}\.\d{1,6}$'
    if re.match(pattern, longitude):
        return longitude
    else:
        return '0.000000'
```

```python
sql_statements = []
special_items = []

poi_id = 300000
poi_res_id = 100000
for item_dict in result_2_csv:
    item = item_dict.to_dict()

    if item['img'] in ['oooo1', 'oooo']:
        special_items.append(item)
        continue

    # 每次循环增加ID
    poi_id += 1
    poi_res_id += 1

    if item['category_id'] == '党群服务中心':
        item['category_id'] = '105'

    # 生成第一个SQL语句
    sql1 = f"""
    INSERT INTO "public"."poi" ("id", "category_id", "name", "address", "city", "county", "longitude", "latitude", "open_time")
    VALUES (
        '{poi_id}', 
        '{item['category_id']}', 
        '{item['title']}', 
        '{item['address']}', 
        '330100',  
        '{get_county_code(item['city'], item['county'])}', 
        '{validate_and_default_longitude(item['longitude'].rstrip(','))}', 
        '{validate_and_default_longitude(item['latitude'].rstrip(','))}', 
        '{item['open_time']}'
    );
    """.replace('\n', '').replace('    ', ' ').strip()
    sql_statements.append(sql1)

    # 生成第二个SQL语句
    sql2 = f"""
    INSERT INTO "public"."poi_resource" ("id", "poi_id", "type", "url") 
    VALUES (
        '{poi_res_id}',   
        '{poi_id}', 
        '2', 
        '{item['img']}'
    );
    """.replace('\n', '').replace('    ', ' ').strip()
    sql_statements.append(sql2)
```
